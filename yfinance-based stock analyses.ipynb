```python
import yfinance as yf
import matplotlib.pyplot as plt
from matplotlib.dates import WeekdayLocator, MonthLocator, YearLocator
import matplotlib.dates as mdates
import pandas as pd

#tickersymbol = '^FCHI' # Symbol for CAC40. Find flere på https://finance.yahoo.com/markets/world-indices/
tickersymbol = '^GDAXI' # Symbol for DAX
tickersymbol = '^NDX'   # Symbol for NASDAQ100
tickersymbol = '^IXIC'  # Symbol for NASDAQ Composite
tickersymbol = '^GSPC'  # Symbol for S&P500
tickersymbol = '^DJI'   # Dow Jones Industrial Av
tickersymbol = '^NYA'   # NYSE COMPOSITE (DJ)
tickersymbol = '^OMXC25'# Copenhagen
tickersymbol = '^OMXH25'# Helsinki
tickersymbol = '^OMX'   # Stockholm

tickerdata = yf.Ticker(tickersymbol)
tickerDf = tickerdata.history(period='1d', start='2025-01-01', end = '2025-02-07')
tickerDf['Date'] = tickerDf.index
tickerDf['Date'] = pd.to_datetime(tickerDf['Date'], format='%m-%d')

yloc = YearLocator()
mloc = MonthLocator()
wloc = WeekdayLocator()
fig, ax = plt.subplots(1, 1)
#ax.plot(tickerDf['Date'], tickerDf['Close'], label = 'CAC40')
ax.plot(tickerDf['Date'], (tickerDf['Close'] - tickerDf['Close'].iloc[0])*100/tickerDf['Close'].iloc[0], label = 'CAC40')

ax.xaxis.set_major_locator(mloc)
#ax.xaxis.set_minor_locator(wloc)

ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%y'))
#ax.xaxis.set_minor_formatter(mdates.DateFormatter('%m-%d'))
ax.grid('both');
ax.legend();
ax.set_xlabel('Date');
ax.set_ylabel('Change [%]');
```
